**Problem Statement:**

This data contains region(district) wise rainfall across India.

1. Import the data into Python environment as a Pandas DataFrame.
2. Check for missing values, if any and drop the corresponding rows.
3. Find the district that gets the highest annual rainfall.
4. Display the top 5 states that get the highest annual rainfall.
5. Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.
6. Display the state-wise mean rainfall for all the months using a pivot table.
7. Display the count of districts in each state.
8. For each state, display the district that gets the highest rainfall in May. Also display the recorded rainfall

```python
'''1.Importing the data'''

import numpy as np
import pandas as pd
df = pd.read_csv('rainfall.csv')
df.head()
```

```python
'''2.Checking for missing values'''
df.info()
```

There are no missing value in data

```python
'''3.checking for highest rainfall value and district'''

max_annual_rainfall_idx = df['ANNUAL'].idxmax()

# Get the district with the highest annual rainfall
district_highest_rainfall = df.loc[max_annual_rainfall_idx, 'DISTRICT']
annual_rainfall = df.loc[max_annual_rainfall_idx, 'ANNUAL']

# Print the result
print(f"The district with the highest annual rainfall is {district_highest_rainfall} with an annual rainfall of {annual_rainfall}.")
```

```python
'''4.To get top 5 states that gets highest rainfall'''

total_annual_rainfall = df.groupby('STATE_UT_NAME')['ANNUAL'].sum()

top_5_states = total_annual_rainfall.sort_values(ascending=False).head()

top_5_states
```

```python
'''5.To Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'''

df.drop(columns=['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'],inplace=True)
df.head()
```

```python
'''6.To display the state-wise mean rainfall for all the months using a pivot table'''

pivot_table = df.pivot_table(index='STATE_UT_NAME',
                             values=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'],
                             aggfunc=np.mean)
pivot_table.reset_index()
```

```python
'''7.Displaying the count of districts in each state'''

group = df.groupby(['STATE_UT_NAME']).count()[['DISTRICT']]
group.reset_index()
```

```python
'''8.For each state, displaying the district that gets the highest rainfall in May. Also display the recorded rainfall'''

max_rainfall_may = df.groupby('STATE_UT_NAME')['MAY'].idxmax() # To find the index of max rainfall in May for each state
result = df.loc[max_rainfall_may, ['STATE_UT_NAME', 'DISTRICT', 'MAY']]
result.reset_index(drop=True) # To rest index and drop old index
```
